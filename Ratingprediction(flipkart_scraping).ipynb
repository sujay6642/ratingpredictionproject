```python
#creating a function to scrape text from a list
def scrape_text(web_el):
    list1=[]
    for i in web_el:
        try:
            
            list1.append(i.text.replace("\n",","))
        except Exception as e:
            print("exception in scrape_text",e)
            list1.append("--")
    return list1
```

```python
def get_attribute(web_el):
    atr=[]
    try:
        for i in web_el:
            atr.append(i.get_attribute('href'))
    except Exception as p:
        print("attribute error:",p)
    return atr
            
        
        
```

```python
#Creating a program to scrap rating and reviews from the extracted urls
count=0

#keys=['most recent']
for i in urls_:
    try:
        
        driver.get(i)
        time.sleep(1)
        #Clicking on the "see all reviews box"
        all_rev=driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']")
        all_rev.click()
        time.sleep(2)
        #clicking on the  comment customization bar
        custom=driver.find_element_by_xpath("//select[@class='_1EDlbo tVKh2S']")
        #for key in keys:
        custom.send_keys('most recent')
        time.sleep(1)
        pages_=driver.find_elements_by_xpath("//a[@class='ge-49M']")
        pages=get_attribute(pages_)
        for k in range(len(pages)):
                
            time.sleep(1)
            #The first character of every row is the rating hence extracting the complete text data 
            rating_el=driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']")
            
            full_text=scrape_text(rating_el)
            time.sleep(1)
                #if(len(rating_el)>1):
            rate=[]
            for j in full_text:
                
                rate.append(j[0])
            rating.append(rate)
            #text_list=full_text.split(",")
            
            review_el=driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
            review_=scrape_text(review_el)
            review.append(review_)
            heading_el=driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
            head_=scrape_text(heading_el)
               
            head.append(head_)
            #time.sleep(1)
            #page=pages[k].get_attribute('href')
            driver.get(pages[k])
        count=count+1    
    except TimeoutException as t:
        print("Time out ",count,t)
        count=count+1
    except StaleElementReferenceException as s:
        print("Stale element",count,s)
        count=count+1
    except NoSuchElementException as nse:
        print("Element not found in",count,nse)
        count=count+1
        head.append("--")
        review.append("--")
        rating.append("--")
    except Exception as e:
        print("Exception raised ",count,e)
        count=count+1
        
print(len(rating))
print(len(review))
print(len(head))

    
```

```python
#Extracting lists in ordered fashion
ratings_2=[]
heading_2=[]
reviews_2=[]
c=0
for i in range(len(head)):
    try:
        for j in range(len(rating[i])):
        
            
            if(head[i]!="--")or(head[i]!=''):
                if(rating[i][0]!='-'):
                    heading_2.append(head[i][j])
            if(review[i]!="--")or(review[i]!=''):
                if(rating[i][0]!='-'):
                    reviews_2.append(review[i][j])
            if(rating[i]!="--")or(rating[i]!=''):
                if(rating[i][0]!='-'):
                    ratings_2.append(rating[i][j])
    except IndexError as e:
        print("index error in: ",c)
    c=c+1     
print(len(ratings_2))
print(len(reviews_2))
print(len(heading_2))
```
